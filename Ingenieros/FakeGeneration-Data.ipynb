```python
import pandas as pd
import random
from faker import Faker

# Inicializar Faker con múltiples idiomas
fake = Faker(['es_ES', 'en_US', 'fr_FR', 'it_IT', 'de_DE'])




# -----------------------------
# 3. COACH TABLE (sin Person)
# -----------------------------
def generate_coach_table(n_coaches=20, seed=42):
    Faker.seed(seed)
    random.seed(seed)
    fake = Faker()

    coaches = []
    for coach_id in range(1, n_coaches + 1):
        coaches.append({
            'coach_id': coach_id,
            'name': fake.name(),
            'birthdate': fake.date_of_birth(minimum_age=35, maximum_age=65),
            'nationality': fake.country(),
            'role': 'Head Coach'
        })
    return pd.DataFrame(coaches)


# -----------------------------
# 4. TEAM TABLE
# -----------------------------


def generate_team_table(df_coach, df_stadium):
    teams_laliga = [
        'Real Madrid', 'Atlético de Madrid', 'Real Betis', 'FC Barcelona', 'Athletic Club',
        'Valencia CF', 'Sevilla FC', 'RCD Espanyol', 'Real Sociedad', 'UD Las Palmas',
        'Real Valladolid', 'RCD Mallorca', 'Celta de Vigo', 'CA Osasuna', 'Villarreal CF',
        'Deportivo Alavés', 'Getafe CF', 'Rayo Vallecano', 'Girona FC', 'CD Leganés'
    ]

    # Asignar un estadio a cada equipo (usaremos el estadio en orden, puedes personalizarlo como quieras)
    stadium_ids = df_stadium['stadium_id'].tolist()

    # Aseguramos que hay suficientes estadios
    if len(stadium_ids) < len(teams_laliga):
        raise ValueError("No hay suficientes estadios en df_stadium para asignar a todos los equipos.")

    # Creamos la lista de datos de equipos
    team_data = []
    for team_id, (team_name, coach) in enumerate(zip(teams_laliga, df_coach.itertuples()), start=1):
        # Asignamos un estadio de manera cíclica o aleatoria
        stadium_id = stadium_ids[team_id % len(stadium_ids)]  # Si tienes más equipos que estadios, asignamos en ciclo
        team_data.append({
            'team_id': team_id,
            'coach_id': coach.coach_id,
            'stadium_id': stadium_id,  # Añadimos el estadio aquí
            'name': team_name,
            'league': 'LaLiga'
        })

    return pd.DataFrame(team_data)


# -----------------------------
# Tabla de Personas (person_table)
# -----------------------------
# Esta tabla contiene información básica sobre las personas (jugadores y entrenadores).
# Cada persona tiene un identificador único (person_id), nombre, fecha de nacimiento (con edad entre 18 y 45 años),
# y nacionalidad, generados de manera aleatoria utilizando la librería Faker para simular datos realistas.

# -----------------------------
# Tabla de Entrenadores (coach_table)
# -----------------------------
# Esta tabla contiene los entrenadores de los equipos. Cada entrenador tiene un identificador único (coach_id)
# y un identificador de persona (person_id) que se relaciona con la tabla de personas. 
# Se asigna aleatoriamente un entrenador a cada equipo, con el rol de "Head Coach".

# -----------------------------
# Tabla de Equipos (team_table)
# -----------------------------
# Esta tabla contiene los equipos de LaLiga. Cada equipo tiene un identificador único (team_id), 
# el identificador de su entrenador (coach_id), el identificador de su estadio (stadium_id), 
# el nombre del equipo y la liga a la que pertenece (en este caso, LaLiga).
# Se asegura que haya un estadio asignado a cada equipo. Si los estadios no son suficientes, se asignan de manera cíclica.

```

```python
import numpy as np
from datetime import datetime, timedelta


# -----------------------------
# 5. PLAYER TABLE (sin Person)
# -----------------------------
def generate_player_table(df_coach, df_team, n_players=100, seed=42):
    Faker.seed(seed)
    random.seed(seed)
    fake = Faker()

    team_ids = df_team['team_id'].tolist()
    num_teams = len(team_ids)

    players_per_team = n_players // num_teams
    extras = n_players % num_teams

    assignments = []
    for team_id in team_ids:
        assignments += [team_id] * players_per_team
    assignments += random.choices(team_ids, k=extras)
    random.shuffle(assignments)

    positions = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
    player_data = []

    for idx, team_id in enumerate(assignments, start=1):
        player_data.append({
            'player_id': idx,
            'name': fake.name(),
            'birthdate': fake.date_of_birth(minimum_age=18, maximum_age=40),
            'nationality': fake.country(),
            'position': random.choice(positions),
            'market_value': round(random.uniform(1_000_000, 100_000_000), 2),
            'current_team_id': team_id,
            'instagram_followers': random.randint(1_000, 10_000_000),
            'engagement_rate': round(random.uniform(0.5, 10.0), 2),
            'media_mentions': random.randint(10, 5000)
        })

    return pd.DataFrame(player_data)



# -----------------------------
# 6. MATCH TABLE
# -----------------------------
import pandas as pd
import random
from datetime import datetime, timedelta

def generate_match_table(df_team, df_stadium, seed=42):
    random.seed(seed)
    team_ids = df_team['team_id'].tolist()
    
    # Crear un diccionario para obtener el nombre del equipo a partir del ID
    team_name_dict = dict(zip(df_team['team_id'], df_team['name']))

    # Crear un diccionario para obtener el nombre del estadio a partir del ID
    stadium_name_dict = dict(zip(df_stadium['stadium_id'], df_stadium['name']))

    if 'capacity' in df_stadium.columns:
        stadium_capacity_dict = dict(zip(df_stadium['stadium_id'], df_stadium['capacity']))
    else:
        stadium_capacity_dict = {sid: random.randint(20000, 80000) for sid in df_stadium['stadium_id']}
    
    matches = []
    match_id = 1
    match_dates = pd.date_range(start='2023-08-01', periods=300).tolist()

    # Generar ida y vuelta para cada par único de equipos
    for i in range(len(team_ids)):
        for j in range(i + 1, len(team_ids)):
            home_1, away_1 = team_ids[i], team_ids[j]
            home_2, away_2 = team_ids[j], team_ids[i]

            for home, away in [(home_1, away_1), (home_2, away_2)]:
                match_date = random.choice(match_dates)
                result = random.choice(['Home Win', 'Away Win', 'Draw'])
                total_time = 90 + random.choice([0, 2, 4, 5, 6])

                # Obtener el nombre del estadio usando el ID del estadio
                stadium_name = stadium_name_dict[home]

                # Obtener el nombre de los equipos
                home_team_name = team_name_dict[home]
                away_team_name = team_name_dict[away]

                stadium_capacity = stadium_capacity_dict[home]
                attendance = random.randint(int(0.5 * stadium_capacity), stadium_capacity)

                matches.append({
                    'match_id': match_id,
                    'stadium_id': home,  # Mantener el ID del estadio
                    'stadium_name': stadium_name,  # Agregar el nombre del estadio
                    'home_team_id': home,
                    'away_team_id': away,
                    'home_team_name': home_team_name,  # Agregar el nombre del equipo local
                    'away_team_name': away_team_name,  # Agregar el nombre del equipo visitante
                    'match_date': match_date.date(),
                    'result': result,
                    'total_time': total_time,
                    'attendance': attendance
                })

                match_id += 1

    return pd.DataFrame(matches)




# -----------------------------
# 7. PLAYER STATS TABLE (con nuevos campos)
# -----------------------------
def generate_player_stats_table(df_player, df_match, n_stats=300, seed=42):
    random.seed(seed)
    stats = []

    for stat_id in range(1, n_stats + 1):
        match = random.choice(df_match['match_id'].tolist())
        player = random.choice(df_player['player_id'].tolist())

        # Para simular días de descanso, vamos a generar un número entre 0 y 14 (por ejemplo)
        rest_days = random.randint(0, 14)

        # Simulamos si el jugador tiene un partido fuera consecutivo (back-to-back away)
        is_back_to_back_away = random.choice([0, 1])

        # Simulamos si ha ocurrido una lesión en el partido (probabilidad baja)
        injury_occurred = 1 if random.random() < 0.33 else 0

        stats.append({
            'stat_id': stat_id,
            'match_id': match,
            'player_id': player,
            'goals': random.randint(0, 3),
            'assists': random.randint(0, 2),
            'yellow_cards': random.choice([0, 1]),
            'red_cards': random.choice([0, 1 if random.random() < 0.05 else 0]),
            'minutes_played': random.randint(30, 95),
            'distance_covered_in_meters': random.randint(2000, 12000),
            'sprints': random.randint(5, 40),
            'pass_completion_rate': random.randint(60, 100),

            # Nuevos campos:
            'injury_occurred': injury_occurred,
            'rest_days_since_last_match': rest_days,
            'is_back_to_back_away_game': is_back_to_back_away
        })

    return pd.DataFrame(stats)



# -----------------------------
# Tabla de Jugadores (player_table)
# -----------------------------
# Esta tabla contiene los jugadores de fútbol. Cada jugador tiene un identificador único (player_id), 
# su identificador de persona (person_id), la posición en la que juega, su valor de mercado aleatorio, 
# el equipo al que pertenece, seguidores en Instagram, tasa de compromiso en redes sociales, 
# y menciones en medios. Los jugadores se distribuyen aleatoriamente entre los equipos disponibles 
# con un reparto equitativo, y los jugadores sobrantes se asignan aleatoriamente.

# -----------------------------
# Tabla de Partidos (match_table)
# -----------------------------
# Esta tabla contiene los partidos entre equipos. Cada partido tiene un identificador único (match_id), 
# los identificadores de los equipos locales y visitantes (home_team_id, away_team_id), 
# el nombre de los equipos (home_team_name, away_team_name), el nombre del estadio (stadium_name), 
# la fecha del partido, el resultado (Home Win, Away Win, Draw), y el tiempo total de juego (total_time). 
# Los partidos se generan de manera que cada par de equipos juegue un partido de ida y vuelta.

# -----------------------------
# Tabla de Estadísticas de Jugadores (player_stats_table)
# -----------------------------
# Esta tabla contiene las estadísticas de cada jugador en los partidos. Cada entrada tiene un identificador único 
# (stat_id), el identificador del partido (match_id), el identificador del jugador (player_id), y diversas estadísticas 
# del rendimiento del jugador en el partido como goles, asistencias, tarjetas, minutos jugados, distancia recorrida, 
# sprints, y porcentaje de pases completados. Se generan aleatoriamente estadísticas de jugadores en distintos partidos.

```

```python

# Generar nuevas tablas
df_player = generate_player_table(df_coach, df_team, n_players=100)
df_match = generate_match_table(df_team, df_stadium)
df_player_stats = generate_player_stats_table(df_player, df_match)

# Ver resultados
print("Df player")
print(df_player.head())
print("Df Match")
print(df_match.head())
print("Df player stats")
print(df_player_stats.head())
```

```python
# -----------------------------
# 8. WEATHER TABLE
# -----------------------------
def generate_weather_table(df_match, seed=42):
    random.seed(seed)
    conditions = ['Sunny', 'Cloudy', 'Rain', 'Snow', 'Fog', 'Windy']
    precipitation_types = ['None', 'Rain', 'Snow', 'Drizzle', 'Hail']

    weather_data = []

    for match in df_match.itertuples():
        weather_data.append({
            'match_id': match.match_id,
            'weather_id': match.match_id,  # Un weather_id por match
            'temperature': round(random.uniform(5.0, 35.0), 2),
            'conditions': random.choice(conditions),
            'humidity': round(random.uniform(30.0, 90.0), 2),
            'precipitation': random.choice(precipitation_types),
            'wind_speed_in_km_h': round(random.uniform(0, 30), 2)
        })

    return pd.DataFrame(weather_data)


# -----------------------------
# 9. FAN TABLE
# -----------------------------
def generate_fan_table_simple(n_fans=50, seed=42):
    random.seed(seed)
    fans = []
    for fan_id in range(1, n_fans + 1):
        fans.append({
            'fan_id': fan_id,
            'loyalty_points': random.randint(0, 1000)
        })
    return pd.DataFrame(fans)




# -----------------------------
# 10. TICKET TABLE
# -----------------------------
def generate_ticket_table(df_fan, df_match, n_tickets=100, seed=42):
    random.seed(seed)
    tickets = []

    for ticket_id in range(1, n_tickets + 1):
        fan_id = random.choice(df_fan['fan_id'].tolist())
        match_id = random.choice(df_match['match_id'].tolist())
        seat_number = f"{random.randint(1, 30)}{random.choice(['A', 'B', 'C', 'D'])}"
        price = round(random.uniform(20, 120), 2)
        purchase_date = fake.date_between(start_date="-6M", end_date="today")

        tickets.append({
            'ticket_id': ticket_id,
            'fan_id': fan_id,
            'match_id': match_id,
            'seat_number': seat_number,
            'price': price,
            'purchase_date': purchase_date
        })

    return pd.DataFrame(tickets)


# -----------------------------
# Tabla de Clima (weather_table)
# -----------------------------
# Esta tabla contiene las condiciones climáticas para cada partido. Cada registro tiene un identificador único (weather_id), 
# el identificador del partido (match_id), la temperatura (temperature), las condiciones climáticas (conditions), 
# la humedad (humidity), el tipo de precipitación (precipitation), y la velocidad del viento (wind_speed_in_km_h). 
# Las condiciones climáticas y otras variables se generan aleatoriamente para cada partido.

# -----------------------------
# Tabla de Aficionados (fan_table)
# -----------------------------
# Esta tabla contiene los aficionados al fútbol. Cada aficionado tiene un identificador único (fan_id), 
# su identificador de persona (person_id), y los puntos de lealtad (loyalty_points). Los aficionados son seleccionados 
# aleatoriamente de la tabla de personas, excluyendo a los jugadores y entrenadores. El número de aficionados se puede 
# ajustar con el parámetro n_fans.

# -----------------------------
# Tabla de Boletos (ticket_table)
# -----------------------------
# Esta tabla contiene los boletos comprados por los aficionados para los partidos. Cada boleto tiene un identificador único 
# (ticket_id), el identificador del aficionado que compró el boleto (fan_id), el identificador del partido (match_id), 
# el número de asiento (seat_number), el precio del boleto (price), y la fecha de compra (purchase_date). Los boletos 
# se generan aleatoriamente con valores de precio y asiento, y se asignan a aficionados y partidos aleatorios.
```

```python
# Generar tablas
df_weather = generate_weather_table(df_match)
df_fan = generate_fan_table_simple(n_fans=50)
df_ticket = generate_ticket_table(df_fan, df_match)

# Visualización rápida
print(df_weather.head())
print(df_fan.head())
print(df_ticket.head())
```

```python
# -----------------------------
# 11. FORMATION TABLE
# -----------------------------
def generate_formation_table(df_coach, n_formations_per_coach=2, seed=42):
    random.seed(seed)
    formation_types = ['4-4-2', '4-3-3', '3-5-2', '5-3-2', '4-2-3-1', '3-4-3']
    formations = []

    formation_id = 1
    for coach in df_coach.itertuples():
        for _ in range(n_formations_per_coach):
            formations.append({
                'formation_id': formation_id,
                'coach_id': coach.coach_id,
                'type': random.choice(formation_types)
            })
            formation_id += 1

    return pd.DataFrame(formations)

# -----------------------------
# Tabla de Formaciones (formation_table)
# -----------------------------
# Esta tabla contiene las formaciones tácticas utilizadas por los entrenadores. Cada registro tiene un identificador único 
# (formation_id), el identificador del entrenador (coach_id), y el tipo de formación táctico (type) usado, que es 
# seleccionado aleatoriamente de un conjunto de formaciones comunes como '4-4-2', '4-3-3', '3-5-2', etc. 
# Cada entrenador puede tener múltiples formaciones, con un número de formaciones por entrenador determinado por el parámetro 
# n_formations_per_coach.


```
